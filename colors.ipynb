```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define a 13-color colorblind-friendly palette
custom_palette = [
    "#000000",  # Black
    "#E69F00",  # Orange
    "#56B4E9",  # Sky Blue
    "#009E73",  # Bluish Green
    "#F0E442",  # Yellow
    "#0072B2",  # Blue
    "#D55E00",  # Vermillion
    "#CC79A7",  # Reddish Purple

    "#999999",  # Grey
    "#FFFFFF",  # White (often used for contrast)
    "#117733",  # Dark Green
    "#882255",  # Wine Red
    "#44AA99",  # Teal
]

# Display the palette visually
sns.palplot(custom_palette)
plt.title("13-Color Colorblind-Friendly Palette", fontsize=14)
plt.show()
```

```python
# Create a 13-color colorblind-friendly palette
palette = sns.color_palette("colorblind", n_colors=13)

# Visualize the palette
sns.palplot(palette)
plt.title("Seaborn Colorblind-Friendly Palette (13 colors)")
plt.show()
```

```python
from colorspace import qualitative_hcl
import seaborn as sns
import matplotlib.pyplot as plt

# Generate 12 hues evenly in HCL space
palette = qualitative_hcl([12], hue_range=(0, 360), chroma=50, luminance=70)

# Optionally simulate how deuteranopic users would see them
#simulated = simulate_cvd(palette, deficiency="deutan")


sns.set_palette(palette)
sns.palplot(palette)
plt.title("12-Color Qualitative HCL Palette")
plt.show()
```
